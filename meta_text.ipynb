#Import Library

```python
from ttsmms import TTS
```

###Update lang code. amh refers to amharic language

```python
tts=TTS("data/amh") 
```

```python
wav=tts.synthesis("የምንጠቀምባቸው?")
```

```python
from IPython.display import Audio
Audio(wav["x"], rate=wav["sampling_rate"])
```

#Example 2

```python
text = """አንድን ድረ-ገጽ ትክክለኛ ወይም ማጭበርበሪያ መሆኑን ለማረጋገጥ የምንጠቀምባቸው ‘ስካምአድቫይዘር’ እና ‘ስካም ዲቴክተር’ ደግሞ ስለ ድርጅቱ ያቀረቡት ትንተና እንመልከት።")"""
wav = tts.synthesis(text)
from IPython.display import Audio
Audio(wav["x"], rate=wav["sampling_rate"])
```

#Download the audio file

```python
# Download the audio file
from google.colab import files
from scipy.io import wavfile
import numpy as np

# Convert audio data to 16-bit signed integer format (assuming it is in the range -1 to 1)
audio_data = np.int16(wav["x"] * 32767)

# Save the audio data as a WAV file
wavfile.write('audio_file.wav', wav["sampling_rate"], audio_data)

# Download the audio file
files.download('audio_file.wav')
```
